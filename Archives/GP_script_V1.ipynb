```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from gp_trajectory_compensation import TrajectoryCompensatorGP
```

```python
# --- Load observed data ---
observed_df = pd.read_csv('experiment_1.csv')
t_sec = observed_df['Time'].values  # (T,)
observed_xyz = observed_df[['x', 'y', 'z']].values  # (T, 3)
T = observed_xyz.shape[0]  # Number of time points in observed data
```

```python
# Commanded waypoints (from G-code)
x_cmd = np.array([-96.839, -58.761, -20.692, 17.378, 55.433, 93.495, 131.553])
y_cmd = np.array([79.866, 79.841, 79.816, 79.791, 79.776, 79.751, 79.726])
z_cmd = np.array([2.91, 3.00, 3.05, 2.95, 2.85, 2.87, 2.89])

# Time vector for the 7 waypoints (you need to define this)
# For example, if the robot takes 7 seconds to move through all waypoints:
t_waypoints = np.linspace(0, 7, 7)  # Adjust start/end time as needed

# Interpolate each axis
x_interp = interp1d(t_waypoints, x_cmd, kind='linear', fill_value='extrapolate')
y_interp = interp1d(t_waypoints, y_cmd, kind='linear', fill_value='extrapolate')
z_interp = interp1d(t_waypoints, z_cmd, kind='linear', fill_value='extrapolate')

# Create commanded_xyz (254195, 3) array
commanded_xyz = np.column_stack([
    x_interp(t_sec),
    y_interp(t_sec),
    z_interp(t_sec)
])
```

```python
# --- Create desired_xyz (T, 3) ---
x_desired = np.linspace(0, 0.3, T)  # Linear from 0 to 0.3
y_desired = np.full(T, 0.9)         # Constant at 0.9
z_desired = np.full(T, 0.8)         # Constant at 0.8
desired_xyz = np.column_stack([x_desired, y_desired, z_desired])
```

```python
# --- Sanity check: All arrays should be (T, 3) ---
print("desired_xyz shape:", desired_xyz.shape)
print("commanded_xyz shape:", commanded_xyz.shape)
print("observed_xyz shape:", observed_xyz.shape)
print("t_sec shape:", t_sec.shape)
```

```python
# --- Fit the GP model ---
comp = TrajectoryCompensatorGP(include_vel=True, dt_hint=0.01, kernel='tiny', damping=1.0)
comp.fit(desired_xyz, commanded_xyz, observed_xyz, t_sec)
```

```python
# --- Predict and compensate ---
predicted_error = comp.predict_error(commanded_xyz, desired_xyz, t_sec)
compensated_cmd = comp.compensate(desired_xyz, t_sec=t_sec, n_iter=3)
```

```python

# Plot x, y, z trajectories
fig, axes = plt.subplots(3, 1, figsize=(12, 8), sharex=True)

# X-axis
axes[0].plot(t_sec, desired_xyz[:, 0], label='Desired x', color='green')
axes[0].plot(t_sec, commanded_xyz[:, 0], label='Commanded x', color='red', linestyle='--')
axes[0].plot(t_sec, observed_xyz[:, 0], label='Observed x', color='blue', alpha=0.7)
axes[0].plot(t_sec, compensated_cmd[:, 0], label='Compensated x', color='purple', linestyle=':')
axes[0].set_ylabel('x (m)')
axes[0].legend()
axes[0].grid(True)

# Y-axis
axes[1].plot(t_sec, desired_xyz[:, 1], label='Desired y', color='green')
axes[1].plot(t_sec, commanded_xyz[:, 1], label='Commanded y', color='red', linestyle='--')
axes[1].plot(t_sec, observed_xyz[:, 1], label='Observed y', color='blue', alpha=0.7)
axes[1].plot(t_sec, compensated_cmd[:, 1], label='Compensated y', color='purple', linestyle=':')
axes[1].set_ylabel('y (m)')
axes[1].legend()
axes[1].grid(True)

# Z-axis
axes[2].plot(t_sec, desired_xyz[:, 2], label='Desired z', color='green')
axes[2].plot(t_sec, commanded_xyz[:, 2], label='Commanded z', color='red', linestyle='--')
axes[2].plot(t_sec, observed_xyz[:, 2], label='Observed z', color='blue', alpha=0.7)
axes[2].plot(t_sec, compensated_cmd[:, 2], label='Compensated z', color='purple', linestyle=':')
axes[2].set_ylabel('z (m)')
axes[2].set_xlabel('Time (s)')
axes[2].legend()
axes[2].grid(True)

plt.suptitle('Trajectory Comparison')
plt.show()
```

```python
def load_and_preprocess_experiments(experiment_names, x_cmd, y_cmd, z_cmd, t_waypoints):
    observed_xyz_list = []
    commanded_xyz_list = []
    desired_xyz_list = []
    t_sec_list = []

    for i, exp_name in enumerate(experiment_names):
        df = pd.read_csv(f'{exp_name}.csv')
        t_sec_i = df['Time'].values
        observed_xyz_i = df[['x', 'y', 'z']].values
        T_i = len(t_sec_i)

        x_interp = interp1d(t_waypoints, x_cmd, kind='linear', fill_value='extrapolate')
        y_interp = interp1d(t_waypoints, y_cmd, kind='linear', fill_value='extrapolate')
        z_interp = interp1d(t_waypoints, z_cmd, kind='linear', fill_value='extrapolate')

        commanded_xyz_i = np.column_stack([
            x_interp(t_sec_i),
            y_interp(t_sec_i),
            z_interp(t_sec_i)
        ])

        x_desired_i = np.linspace(0, 0.3, T_i)
        y_desired_i = np.full(T_i, 0.9)
        z_desired_i = np.full(T_i, 0.8)
        desired_xyz_i = np.column_stack([x_desired_i, y_desired_i, z_desired_i])

        observed_xyz_list.append(observed_xyz_i)
        commanded_xyz_list.append(commanded_xyz_i)
        desired_xyz_list.append(desired_xyz_i)
        t_sec_list.append(t_sec_i + i * max(t_sec_i))

    observed_xyz_all = np.concatenate(observed_xyz_list, axis=0)
    commanded_xyz_all = np.concatenate(commanded_xyz_list, axis=0)
    desired_xyz_all = np.concatenate(desired_xyz_list, axis=0)
    t_sec_all = np.concatenate(t_sec_list, axis=0)

    return (
        observed_xyz_all, commanded_xyz_all, desired_xyz_all, t_sec_all,
        observed_xyz_list, commanded_xyz_list, desired_xyz_list, t_sec_list
    )
```

```python
experiment_names = ["experiment_2", "experiment_3"]
(
    observed_xyz_all, commanded_xyz_all, desired_xyz_all, t_sec_all,
    observed_xyz_list, commanded_xyz_list, desired_xyz_list, t_sec_list
) = load_and_preprocess_experiments(experiment_names, x_cmd, y_cmd, z_cmd, t_waypoints)
```

```python
comp = TrajectoryCompensatorGP(include_vel=True, dt_hint=0.01, kernel='tiny', damping=1.0)
comp.fit(desired_xyz_all, commanded_xyz_all, observed_xyz_all, t_sec_all)
```

```python
# Predict and compensate for the concatenated data
predicted_error_all = comp.predict_error(commanded_xyz_all, desired_xyz_all, t_sec_all)
compensated_cmd_all = comp.compensate(desired_xyz_all, t_sec=t_sec_all, n_iter=3)
```

```python
# Split the compensated command
cumulative_lengths = np.cumsum([len(t) for t in t_sec_list])
compensated_cmd_list = np.split(compensated_cmd_all, cumulative_lengths[:-1])
```

```python
# Visualize the results for each experiment
for i, exp_name in enumerate(experiment_names):
    df = pd.read_csv(f'{exp_name}.csv')
    t_sec_i = df['Time'].values
    observed_xyz_i = df[['x', 'y', 'z']].values
    commanded_xyz_i = commanded_xyz_list[i]
    desired_xyz_i = desired_xyz_list[i]
    compensated_cmd_i = compensated_cmd_list[i]

    # Plot x, y, z trajectories
    fig, axes = plt.subplots(3, 1, figsize=(12, 8), sharex=True)

    # X-axis
    axes[0].plot(t_sec_i, desired_xyz_i[:, 0], label='Desired x', color='green')
    axes[0].plot(t_sec_i, commanded_xyz_i[:, 0], label='Commanded x', color='red', linestyle='--')
    axes[0].plot(t_sec_i, observed_xyz_i[:, 0], label='Observed x', color='blue', alpha=0.7)
    axes[0].plot(t_sec_i, compensated_cmd_i[:, 0], label='Compensated x', color='purple', linestyle=':')
    axes[0].set_ylabel('x (m)')
    axes[0].legend()
    axes[0].grid(True)
    axes[0].set_title(f'Experiment {exp_name}')

    # Y-axis
    axes[1].plot(t_sec_i, desired_xyz_i[:, 1], label='Desired y', color='green')
    axes[1].plot(t_sec_i, commanded_xyz_i[:, 1], label='Commanded y', color='red', linestyle='--')
    axes[1].plot(t_sec_i, observed_xyz_i[:, 1], label='Observed y', color='blue', alpha=0.7)
    axes[1].plot(t_sec_i, compensated_cmd_i[:, 1], label='Compensated y', color='purple', linestyle=':')
    axes[1].set_ylabel('y (m)')
    axes[1].legend()
    axes[1].grid(True)

    # Z-axis
    axes[2].plot(t_sec_i, desired_xyz_i[:, 2], label='Desired z', color='green')
    axes[2].plot(t_sec_i, commanded_xyz_i[:, 2], label='Commanded z', color='red', linestyle='--')
    axes[2].plot(t_sec_i, observed_xyz_i[:, 2], label='Observed z', color='blue', alpha=0.7)
    axes[2].plot(t_sec_i, compensated_cmd_i[:, 2], label='Compensated z', color='purple', linestyle=':')
    axes[2].set_ylabel('z (m)')
    axes[2].set_xlabel('Time (s)')
    axes[2].legend()
    axes[2].grid(True)

    plt.suptitle(f'Trajectory Comparison for {exp_name}')
    plt.tight_layout()
    plt.show()
```
